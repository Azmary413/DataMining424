Importing the necessary libraries

```python
import pandas as pd
import numpy as np
```

Data Collection and Processing

```python
df = pd.read_csv("/content/heart_disease.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop("education",axis=1,inplace=True)
df.drop("BPMeds",axis=1,inplace=True)
df.drop("totChol",axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['glucose'].describe()
```

```python
df['BMI'].describe()
```

```python
df['glucose'] = df['glucose'].fillna(df['glucose'].mean())
df['cigsPerDay'] = df['cigsPerDay'].fillna(df['cigsPerDay'].mean())
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())
df['heartRate'] = df['heartRate'].fillna(df['heartRate'].mean())
```

```python
df.isnull().sum()
```

```python
df['age'].describe()
```

Data transformation

```python
df['Label'] = pd.cut(x=df['age'], bins=[0, 35, 49, 70],
labels=['young', 'Adult','Elderly'])
```

```python
df
```

```python
print("Categories: ")
print(df['Label'].value_counts())
```

```python
dum = pd.get_dummies(df["Label"])
dum
```

```python
merge = pd.concat([df,dum], axis = "columns")
merge
```

```python
final = merge.drop(["Label"], axis = "columns")
final
```
